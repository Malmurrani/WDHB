# The impact of Covid19 on the selected acute hospital admissions

Data from the national data warehouse was downloaded using Business Object for the period between 1/1/2017 and 31/07/2020.
Specific disease codes that usually been admitted to the hospital as acute cases are used for the selection.
These diseases are; IHD, asthma, respiratory infection and pneumonia, cellulitis, heart failure, Stroke, rheumatic fever.

```python
# Using graph_objects
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
from openpyxl import load_workbook
import plotly.express as px
import pandas as pd
df = pd.read_csv('F:/Seagate Backup Plus Drive/Covid19/CovidImpactUptoAugcsv.csv')
#pd.set_option('display.max_columns', None)
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df=df.loc[df['Date'].between('2016-12-31','2020-08-01', inclusive=False)]
dfDate = df.groupby(['Date']).agg({'Number' : 'sum'})
#flattened
dfDate = pd.DataFrame(dfDate.to_records())
#dfDate['Date'] = pd.to_datetime(dfDate['Date'], format='%d/%m/%Y')
dfDate=dfDate.sort_values(by='Date')
fig = px.line(dfDate, x='Date', y="Number")
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
fig.show()
#df.tail()
```

From the above graph there are clear reductions in the daily number of admissions around January each year for the whole study period, except January 2020 there was no clear decrease. However, there were very sharp declines in the daily number between March and early May 2020.  

To make the analysis easier to read, data is grouped on a weekly bases starting the week on Monday.

```python
#df.dtypes weekly start on Monday
df_Week = df.groupby(['Combine', pd.Grouper(key='Date', freq='W-MON')])[['Number']].sum()
#Flattend file
df_Week = pd.DataFrame(df_Week.to_records())
#df_Week.head()
# produces Pandas Series to remember
#data.groupby('month')['duration'].sum() 
# Produces Pandas DataFrame
#data.groupby('month')[['duration']].sum()

df_Week=df_Week.sort_values(by='Date')
fig = px.line(df_Week, x='Date', y="Number")
fig.show()
```

The graph above shows the weekly variability in hospital admissions. There are seasonal cycles with the lowest admissions around December-January each year.  However, there were very distinctive reductions in the weekly hospital admissions between March and June 2020 (lockdown period).

```python
#weekly by ethnicity
import plotly.express as px
Week_Ethnic = df.groupby(['Ethnicity4', pd.Grouper(key='Date', freq='W-MON')])[['Number']].sum()
#Flattend file
Week_Ethnic = pd.DataFrame(Week_Ethnic.to_records())
Week_Ethnic['Date'] = pd.to_datetime(Week_Ethnic['Date'], format='%d/%m/%Y')
Week_Ethnic=Week_Ethnic.sort_values(['Date', 'Ethnicity4'], ascending=[True, True])
Week_Ethnic=Week_Ethnic.pivot(index='Date', columns='Ethnicity4', values='Number')
#df = px.data.stocks(indexed=True)-1
fig = px.area(Week_Ethnic, facet_col="Ethnicity4", facet_col_wrap=2)
fig.show()
```

Ethnicity groups are shown in the above graph, Maori, Pacific and the Others show the same previous pattern, decline during the lockdown period.  However, Asian ethniciy showed no clear pattern.

```python
#weekly by diagnosis
import plotly.express as px
Week_Diag = df.groupby(['DiagnosisName', pd.Grouper(key='Date', freq='W-MON')])[['Number']].sum()
#Flattend file
Week_Diag = pd.DataFrame(Week_Diag.to_records())
Week_Diag['Date'] = pd.to_datetime(Week_Diag['Date'], format='%d/%m/%Y')
Week_Diag=Week_Diag.sort_values(['Date', 'DiagnosisName'], ascending=[True, True])
Week_Diag=Week_Diag.pivot(index='Date', columns='DiagnosisName', values='Number')
#df = px.data.stocks(indexed=True)-1
fig = px.area(Week_Diag, facet_col="DiagnosisName", facet_col_wrap=2)
fig.show()
```

The diagnosis of the conditions admitted to hospital shown on the above graph. cellulitis, asthema, heart failure and pneumonia declined during the lockdown period. That could be explained by the ability to treat them in the community or there was actual decline in the incidence of those diseases.  However, IHD admissions also declined with no clear explanation and this raise concern. 

```python
#weekly by Age
import plotly.express as px
Week_Age = df.groupby(['Age_Group4', pd.Grouper(key='Date', freq='W-MON')])[['Number']].sum()
#Flattend file
Week_Age = pd.DataFrame(Week_Age.to_records())
Week_Age['Date'] = pd.to_datetime(Week_Age['Date'], format='%d/%m/%Y')
Week_Age=Week_Age.sort_values(['Date', 'Age_Group4'], ascending=[True, True])
Week_Age=Week_Age.pivot(index='Date', columns='Age_Group4', values='Number')
#df = px.data.stocks(indexed=True)-1
fig = px.area(Week_Age, facet_col="Age_Group4", facet_col_wrap=2)
fig.show()
```

All age groups were affected during the lockdown period as shown in the graph above.

In conclusion there is clear impact on the hospital acute admissions for the selected conditions.  will monitor these diseases through time to identify any change in the trend.
We need to understand why IHD declined during the lockdown period.
